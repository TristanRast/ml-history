# McCulloch-Pitts Neuron

* Inputs: binary (0 or 1)
* Weights: integers
* Threshold: if weighted sum >= threshold, fire (1), else don't (0).

```python
# inputs for weight and threshold
class MCPNeuron:
    def __init__(self, weights, threshold):
        self.weights = weights
        self.threshold = threshold

# Activation based on threshold of init
    def activate(self, inputs):
        # sums the weights * their associated input
        weighted_sum = sum(w * x for w, x in zip(self.weights, inputs))
        return 1 if weighted_sum >= self.threshold else 0

# AND gaet: both inputs must be 1
and_gate = MCPNeuron(weights=[1, 1], threshold = 2)

# OR gate: at least one input must be 1
or_gate = MCPNeuron(weights = [1, 1], threshold = 1)

# NOT gate: single input, inverts the logic
not_gate = MCPNeuron(weights = [-1], threshold = 0)
```

# Gates
Using the above gates and the activation logic, we now have the fundamental building blocks that allow for computation.

Use the below truth gate test block to see the outputs that the MCPNeuron provides based on an array of inputs

```python
print("AND Gate")
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(x1, x2, "->", and_gate.activate([x1, x2]))

print("\nOR Gate")
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(x1, x2, "->", or_gate.activate([x1, x2]))

print("\nNot Gate")
for x in [0, 1]:
    print(x, "->", not_gate.activate([x]))
```
